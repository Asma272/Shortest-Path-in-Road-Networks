```python
import gzip
import networkx as nx
```

```python
def load_partial_graph(file_path, max_edges):
    G = nx.Graph()
    edge_count = 0

    with open(file_path, "r") as file: 
        for line in file:
            if line.startswith("#"):
                continue

            u, v = map(int, line.split())
            G.add_edge(u, v, weight=1)
            edge_count += 1

            if edge_count >= max_edges:
                break

    return G
```

```python
file_path = "roadNet-CA.txt"
```

```python
G_small = load_partial_graph(file_path, 1000)
G_medium = load_partial_graph(file_path, 10000)
G_large = load_partial_graph(file_path, 50000)
```

```python
print("Small Graph:")
print("Number of nodes:", G_small.number_of_nodes())
print("Number of edges:", G_small.number_of_edges())

print("\nMedium Graph:")
print("Number of nodes:", G_medium.number_of_nodes())
print("Number of edges:", G_medium.number_of_edges())

print("\nLarge Graph:")
print("Number of nodes:", G_large.number_of_nodes())
print("Number of edges:", G_large.number_of_edges())
```

```python
print("Sample edges from Small Graph:")
edges_list = list(G_small.edges(data=True))[:10]
for edge in edges_list:
    print(edge)
```
